```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

Data is very unbalanced as Fraudlent Transactions are only 492 and rest are the otherwise. So we have to balanced the data before analyzing

```python
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print(legit.shape, fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the vales for both

df.groupby('Class').mean()
```

Cleaning the data using **under-sampling** method

Build a sample dataset containing similar distribution of both transactions.

```python
# Select random values
legit_sample = legit.sample(n=492)

```

```python
# Concatenating

dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
dataset.head()
```

```python
dataset.Class.value_counts()
```

```python
dataset.groupby('Class').mean()
```

```python
X = dataset.drop(columns='Class', axis=1)
Y = dataset.Class
```

```python
print(X.shape, Y.shape)
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_Test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
train_prediction = model.predict(X_train)
train_accuracy = accuracy_score(train_prediction, Y_train)
print('Accuracy on Training data : ', train_accuracy)
```

```python
test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(test_prediction, Y_Test)
print('Accuracy on Testing data : ', test_accuracy)
```
